```python
# String to Integer (ATOI)

""" LEETCODE Q-8. MEDIUM """
def myAtoi(s):
    ans = 0
    sign = 0
    check = True
    signCheck = True
    i = 0
    while check == True and i < len(s):

        if s[i] == " ":
            if signCheck == False:
                check = False
            i += 1
        elif ord(s[i]) >= 48 and ord(s[i]) <= 57:
            ans = ans*10 + int(s[i])
            signCheck = False
            i += 1

        elif s[i] == "-" or s[i] == "+":
            if signCheck:
                if s[i] == "-":
                    sign += 1
            else:
                check = False

            signCheck = False
            i += 1
        else:
            check = False
            i += 1

    if sign%2 == 1:
        if -ans > -(2**31):
            return(-ans)
        else:
            return -((2)**31)
    else:
        if ans < (2**31) - 1:
            return(ans)
        else:
            return 2**31 - 1
```

```python
# Printing all possible subsequences
""" 
Subsequence is basically similar to subarray the only difference being that a subsequence can conatain elements that
        are non continuous, like in subarray.
        
There are basically 2 approaches to print all possible subsequences..
        1. Recursion
        2. Power Set
    
1. Recursion:
    simply an approach where we basically write all possible subarrays, but with a twist. The twist being that 
        while creating a subarray for each element we will add two cases- we either consider it or we don't
        this helps in getting an possible continuous and non-continuous subsequences. 
        """

# a = [1,2,3,4,5,6,7]
# i = 0

def isPalindrome(l):
    a = list(l)
    a.reverse()
    if a == list(l):
        return True
    else:
        return False

def allSubseqences(i, l, arr, ans):
    
    if i >= len(arr) - 1:
        print("L: ", l)
        if isPalindrome(l):
            print("Pal:", l, len(ans))
            if len(l) > len(ans):
                ans = l
        return ans

    
    l.append(arr[i])
    ans = allSubseqences(i + 1, l, arr, ans)
    l.remove(arr[i])
    ans = allSubseqences(i + 1, l, arr, ans)
    

    return ans

if __name__ == "__main__":
    l = [1,2,3,2,1,4,5,6]
    print((allSubseqences(0, [], l, [])))
```
